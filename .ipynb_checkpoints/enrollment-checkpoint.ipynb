```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import figure

#Data from here: https://www.ibhe.org/enrollment.html
```

```python
data_20 = pd.read_csv('data/2020.txt')
data_16 = pd.read_csv('data/2016.txt')
data_15 = pd.read_csv('data/2015.txt')
```

```python
UG20 = data_20.loc[(data_20['UG-head'] > 0)]['UG-head'] 
UG16 = data_16.loc[(data_16['UG-head'] > 0)]['UG-head'] 
UG15 = data_15.loc[(data_15['UG-head'] > 0)]['UG-head'] 

GR20 = data_20.loc[(data_20['UG-head'] > 0)]['GR-head'] 
GR16 = data_16.loc[(data_16['UG-head'] > 0)]['GR-head'] 
GR15 = data_15.loc[(data_15['UG-head'] > 0)]['GR-head'] 

tot20 = UG20 + GR20
tot16 = UG16 + GR16
tot15 = UG15 + GR15

college = data_20.loc[(data_20['UG-head'] > 0)]['Institution'] 

print(len(tot20))
print(len(tot16))
print(len(tot15))
print(len(college))

#college15.to_csv('out15.csv')
#college16.to_csv('out16.csv')
#college20.to_csv('out20.csv')
```

```python
tot20_1 = tot20[0:30]
tot16_1 = tot16[0:30]
tot15_1 = tot15[0:30]
college_1 = college[0:30]

tot20_2 = tot20[31:len(college)]
tot16_2 = tot16[31:len(college)]
tot15_2 = tot15[31:len(college)]
college_2 = college[31:len(college)]
```

```python
fig, ax = plt.subplots(figsize=(20,20))
xticks = college_1
bar_width = 0.2
opacity = 0.8
index = np.arange(len(college_1))


rects1 = plt.bar(index, tot15_1, bar_width,
alpha=opacity,
color='r', label='2015')

rects2 = plt.bar(index+bar_width, tot16_1, bar_width,
alpha=opacity,
color='b', label='2016')

rects3 = plt.bar(index+(2*bar_width), tot20_1, bar_width,
alpha=opacity,
color='g', label='2020')


plt.xticks(index + bar_width, college)

plt.xticks(rotation=90)

plt.ylabel('Total Enrollment')
#plt.title('UG Enrollment')
plt.legend()

plt.tight_layout()
plt.savefig("plot1.pdf")
```

```python
fig, ax = plt.subplots(figsize=(20,20))
xticks = college_2
bar_width = 0.2
opacity = 0.8
index = np.arange(len(college_2))


rects1 = plt.bar(index, tot15_2, bar_width,
alpha=opacity,
color='r', label='2015')

rects2 = plt.bar(index+bar_width, tot16_2, bar_width,
alpha=opacity,
color='b', label='2016')

rects3 = plt.bar(index+(2*bar_width), tot20_2, bar_width,
alpha=opacity,
color='g', label='2020')


plt.xticks(index + bar_width, college_2)

plt.xticks(rotation=90)

plt.ylabel('Total Enrollment')
#plt.title('UG Enrollment')
plt.legend()

plt.tight_layout()
plt.savefig("plot2.pdf")
```
